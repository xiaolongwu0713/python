### To simulate the CC between two random singals resembling brain signals in [0,3] Hz.
question: how to properly simulate such brain signals.
Two important things:

1, If you want to consider all frequencies between [0,3], obey the 1/f law;

2, If just use one frequency will be used, such as 2Hz, is it a good approximation?

A better way, you can just randomly select part of data from EEG data and do CC.


### Method 1: generate random signals.
Discard: not a good approximation.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def _randow_wave(frequency,phase, sr,duration):
    start_time = 0
    time = np.arange(start_time, duration, 1/sr)
    amplitude = 1
    sinewave = amplitude * np.sin(2 * np.pi * frequency * time + phase)
    return sinewave
    
```

```python
for i in range(5000):
    random.uniform(0, 3)
    a=_randow_wave()
```

```python
plt.plot(_randow_wave(1,np.pi/2,1000,1))
plt.show()
```
