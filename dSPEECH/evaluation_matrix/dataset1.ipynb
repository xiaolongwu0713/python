Public SEEG dataset; Number of subjects: 10;

There are 20 trials for subject 6. This script first export 20 individual trial (pred and target) to wave file.


```python
from dSPEECH.config import *
import numpy as np
import matplotlib.pyplot as plt
import librosa
```

```python
folder=data_dir+'evaluation_matrix/dataset1/reconstructed/mel_23/sid6/'
file_target='waveform_truth.wav'
file_pred='waveform_pred.wav'
```

```python
target,sf=librosa.load(folder+file_target)
pred,sf=librosa.load(folder+file_pred)
```

```python
%matplotlib inline
plt.plot(target)
plt.plot(pred)
```

```python
trial_num=20
trial_len=int(len(target)/trial_num)
lines=[trial_len*i for i in range(trial_num)]
```

```python
fig,ax=plt.subplots()
ax.plot(target)
for line in lines:
    ax.axvline(x=line, color='green', linestyle='--')
```

```python
trials_target_start=lines
trials_pred_start=trials_target_start
```

```python
import soundfile

for trial in range(trial_num):
    tmp=target[trials_target_start[trial]:trials_target_start[trial]+trial_len]
    soundfile.write(folder+'target_trial_'+str(trial)+'.wav', tmp, sf)
    
    tmp=pred[trials_target_start[trial]:trials_target_start[trial]+trial_len]
    soundfile.write(folder+'pred_trial_'+str(trial)+'.wav', tmp, sf)
    
```
