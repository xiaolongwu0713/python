One target file + one recon file.

1 whole sentence.

```python
from dSPEECH.config import *
import matplotlib.pyplot as plt
import librosa
folder=data_dir+'evaluation_matrix/dataset10/'
filename=folder+'original.wav'
target,sf=librosa.load(filename) # 

filename=folder+'DNN_voc.wav'
pred,sf=librosa.load(filename) # 
```

```python
plt.plot(target)
plt.plot(pred)
```

```python
starts=[int(0.96*sf), int(3.8*sf)]
trial_len=[int(2.6*sf),int(2.25*sf)]
```

```python
import soundfile
for trial in range(len(starts)):
    tmp1=target[starts[trial]:starts[trial]+trial_len[trial]]
    soundfile.write(folder+'target_trial_'+str(trial)+'.wav', tmp1, sf)
    
    tmp2=pred[starts[trial]:starts[trial]+trial_len[trial]]
    soundfile.write(folder+'pred_trial_'+str(trial)+'.wav', tmp2, sf)
```
