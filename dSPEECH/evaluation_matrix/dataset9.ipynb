```python
starts_target=[272604,682403,1076723,1428421,]
starts_pred=[173335,583134,977454,1365225,]
trial_len=[int(3*sf) for _ in range(len(starts_pred))]
trial_len[-1]=int(1.5*sf)
```

```python
import soundfile
for trial in range(len(starts_pred)):
    tmp1=audio[starts_target[trial]:starts_target[trial]+trial_len[trial]]
    soundfile.write(folder+'target_trial_'+str(trial)+'.wav', tmp1, sf)
    
    tmp2=audio[starts_pred[trial]:starts_pred[trial]+trial_len[trial]]
    soundfile.write(folder+'pred_trial_'+str(trial)+'.wav', tmp2, sf)
```
