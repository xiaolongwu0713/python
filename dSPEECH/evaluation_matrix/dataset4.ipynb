```python
from dSPEECH.config import *
import librosa
import matplotlib.pyplot as plt
```

```python
folder=data_dir+'evaluation_matrix/dataset4/'
file_original=['RIGHT_gt.wav','LEFT_gt.wav','LD_gt.wav','HB_gt.wav']
file_reconstrcted=['RIGHT_pred.wav','LEFT_pred.wav','LD_pred.wav','HB_pred.wav']
```

```python
target1,sf=librosa.load(folder+file_original[1])
pred1,sf=librosa.load(folder+file_reconstrcted[1])

plt.plot(target1)
plt.plot(pred1)
```

```python
target2,sf=librosa.load(folder+file_original[1])
pred2,sf=librosa.load(folder+file_reconstrcted[2])

plt.plot(target2)
plt.plot(pred2)
```

```python
target3,sf=librosa.load(folder+file_original[3])
pred3,sf=librosa.load(folder+file_reconstrcted[3])

plt.plot(target3)
plt.plot(pred3)
```

```python
target=target3
pred=pred3
```

```python
trial_num=7
trial_len=len(target)/trial_num
lines=[int(trial_len*i) for i in range(trial_num)]
```

```python
lines[-1]=lines[-1]
lines[-2]=lines[-2]
fig,ax=plt.subplots()
ax.plot(target)
for line in lines:
    ax.axvline(x=line, color='green', linestyle='--')
```

```python
import soundfile
trial_starts=lines
for trial in range(trial_num):
    if trial==trial_num-1:
        tmp1=target[trial_starts[trial]:]
        soundfile.write(folder+'target_trial_'+str(trial)+'.wav', tmp1, sf)
        
        tmp2=pred[trial_starts[trial]:]
        soundfile.write(folder+'pred_trial_'+str(trial)+'.wav', tmp2, sf)
        
    else:
        tmp1=target[trial_starts[trial]:trial_starts[trial+1]]
        soundfile.write(folder+'target_trial_'+str(trial)+'.wav', tmp1, sf)
        
        tmp2=pred[trial_starts[trial]:trial_starts[trial+1]]
        soundfile.write(folder+'pred_trial_'+str(trial)+'.wav', tmp2, sf)
```
